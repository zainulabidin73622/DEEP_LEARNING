```python
def sigmoid(x, w=1, b=0):
    return 1 / (1 + np.exp(-(w * x + b)))

print("✅ [2] Sigmoid Function Defined: sigmoid(x, w=1, b=0)")
```

```python
x = np.linspace(-10, 10, 400)
print("✅ [3] x-values Generated for Plotting from -10 to 10")
```

```python
w_values = [0.5, 0.5, 0.5]  # Steepness values
b_values = [0, 0, 0]    # Position shift values
colors = ['b', 'orange', 'g']  # Colors for lines

print(f"✅ [4] Parameter Sets:")
print(f"  - w (steepness): {w_values}")
print(f"  - b (position): {b_values}\n")
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)

for i, w in enumerate(w_values):
    plt.plot(x, sigmoid(x, w=w, b=0), color=colors[i], label=f'w = {w}, b = 0', linestyle='--')

plt.title('Effect of w (Steepness) on Sigmoid')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid()

print("✅ [5] Plotted Sigmoid for Different w (Steepness)")
```

```python
plt.subplot(1, 2, 2)

for i, b in enumerate(b_values):
    plt.plot(x, sigmoid(x, w=1, b=b), color=colors[i], label=f'w = 1, b = {b}', linestyle='dotted')

plt.title('Effect of b (Position) on Sigmoid')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.legend()
plt.grid()

print("✅ [6] Plotted Sigmoid for Different b (Position)")
```

```python
plt.tight_layout()
plt.show()

print("✅ [7] All Plots Displayed Successfully")
```
